```python
# Filter for complete games (CG >= 1)
complete_games_df = pitching_df[pitching_df['CG'] >= 1]

# Select relevant columns: 'Name' and 'Tm' (Team)
pitchers_with_cg_df = complete_games_df[['Name', 'Tm']]

# Count the number of complete games per pitcher-team combination (optional)
pitchers_with_cg_df = pitchers_with_cg_df.groupby(['Name', 'Tm']).size().to_frame(name='CGs').reset_index()

# Print the DataFrame
print(pitchers_with_cg_df)
```
